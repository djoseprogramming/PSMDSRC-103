```python
###### FUNCTION DEFINITIONS #####
def balance(products): # Show inventory balance
    if len(products) == 0:
        print('No inventory available. Add a new product first.')
    else:
        for x in products:
            print(f"Name: {x['name']}, Qty: {x['qty']}, Price: {x['price']}")

def find_max_price(products): # Task 1 requirement
    if len(products) == 0:
        print('No inventory available. Add a new product first.')
    else:
        print('Products with highest price:')
        max_price = -1
        for x in products:
            if x['price'] > max_price:
                max_price = x['price']
        for x in products:
            if x['price'] == max_price:
                print(f"{x['name']}: {x['price']}")

def find_min_stock(products): # Task 2 requirement
    if len(products) == 0:
        print('No inventory available. Add a new product first.')
    else:
        print('Products with lowest stock:')
        minStock = products[0]['qty']
        for x in products:
            if x['qty'] < minStock:
                minStock = x['qty']
        for x in products:
            if x['qty'] == minStock:
                print(f'{x['name']}: {x['qty']}')

def average_price(products): # Task 3 requirement
    if len(products) == 0:
        print('No inventory available. Add a new product first.')
    else:
        print('Average price selected.')
        total_price = 0
        count = 0
        for x in products:
            if x['qty'] != 0:
                total_price += x['price']
                count += 1
        if count != 0:
            print(f'The average inventory price is: {total_price / count}')
        else:
            print('No products to calculate average price.')
    
def modify_stock(products):
    if len(products) == 0:
        print('No inventory available. Add a new product first.')
    else:
        while True:
            for x in products:
                print(f"Name: {x['name']}, Qty: {x['qty']}, Price: {x['price']}")
                print('')
    
            selection_name = input('Enter the product name to modify:')
            print()
            print('Choose a product detail to modify:')
            print('1 Name')
            print('2 Qty')
            print('3 Price')
            print('0 Back')
            print()
    
            selection_detail = int(input('Enter the product detail to modify:'))
    
            for x in products:
                if x['name'] == selection_name:
                    if selection_detail == 1:
                        print('Name modification selected.')
                        x['name'] = input('Enter the new product name:')
                        print()
                        print(f"Product '{selection_name}' name updated to {x['name']}")
                        
                    elif selection_detail == 2:
                        print('Qty modification selected.')
                        x['qty'] = int(input('Enter the new quantity:'))
                        print()
                        print(f"Product '{selection_name}' qty updated to {x['qty']}")
                        
                    elif selection_detail == 3:
                        print('Price modification selected.')
                        x['price'] = float(input('Enter the new price:'))
                        print()
                        print(f"Product '{selection_name}' price updated to {x['price']}")
                        
                    elif selection_detail == 0:
                        break
                        
                    else:
                        print('Invalid selection.')
            products = remove_out_of_stock(products) # Refresh inventory if zero qty modification was entered
            break
    return products
    
def remove_out_of_stock(products): # Task 4 requirement
    new_inventory = []
    for x in products:
        if x['qty'] == 0:
            print(f"Removed product: {x['name']}, out of stock.")
        else:
            new_inventory.append(x)
    return new_inventory

def new_product(): # New product
    print('Input new product details.')
    name = input('Name:')
    qty = int(input('Quantity:'))
    price = float(input('Price:'))
    print()
    print(f"New product added: {name}, Qty: {qty}, Price: {price}")
    return {'name':name, 'qty':qty, 'price':price}

def main(): # Defines the main program. Task 5
    inventory = []
    while True:
        print()
        print('Inventory Management System')
        print()
        print('1 View inventory balance')
        print('2 Add new product')
        print('3 Find max price')
        print('4 Find average price')
        print('5 Find lowest stock')
        print('6 Modify stock')
        print()
        selection = int(input('Selection: '))
        print()
        if selection == 1:
            balance(inventory)
        elif selection == 2:
            inventory.append(new_product())
        elif selection == 3:
            find_max_price(inventory)
        elif selection == 4:
            average_price(inventory)
        elif selection == 5:
            find_min_stock(inventory)
        elif selection == 6:
            inventory = modify_stock(inventory)
            balance(inventory)
        else:
            print('Invalid Selection.')

        input('Press any key to continue.')
        print("---------------------------------------------------")
```
