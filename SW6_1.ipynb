### 1.

```python
import pandas as pd

# Import the file
filepath = '/content/2019_Yellow_Taxi_Trip_Data.csv'
taxi_df = pd.read_csv(filepath)

# Examine the first 5 rows
taxi_df.head()
```

#### 2.

```python
# Find the dimensions
print(f'''Row: {taxi_df.shape[0]}
Col: {taxi_df.shape[1]}''')
```

#### 3.

```python
# Create a class for repeating summary stats
class summary():

    def __init__(self, df):
        self.mean = df.mean()
        self.median = df.median()
        self.mode = df.mode()
        self.quantile = df.quantile([0.01, 0.05, 0.5, 0.95, 0.99])
        self.min = df.min()
        self.max = df.max()

    def printDetails(self, title):
        print(f'{title.upper()} SUMMARY')
        print(f'Mean: {self.mean}')
        print(f'Median: {self.median}')
        print(f'Mode: {self.mode}')
        print(f'Quantile: {self.quantile}')
        print(f'Min: {self.min}')
        print(f'Max: {self.max}')
        print()

summary(taxi_df['fare_amount']).printDetails("Fare Amount")
summary(taxi_df['tip_amount']).printDetails("Tip Amount")
summary(taxi_df['tolls_amount']).printDetails("Tolls Amount")
summary(taxi_df['total_amount']).printDetails("Total Amount")
```

#### 4.

```python
# Get the longest distance and provide stats
longdist = taxi_df[taxi_df['trip_distance'] == taxi_df['trip_distance'].max()]
longdist[['fare_amount', 'tip_amount', 'tolls_amount', 'total_amount']]
```
